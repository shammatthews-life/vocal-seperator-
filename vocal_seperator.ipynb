```python
import os
import subprocess

# Paste YouTube URL here
YOUTUBE_URL = "https://youtu.be/4NRXx6U8ABQ?si=saT7gvFI3MdR8eZl"

audio_path = "song_from_youtube.mp3"

# Download best audio as MP3
subprocess.run([
    "yt-dlp",
    "-x",
    "--audio-format", "mp3",
    "--audio-quality", "192K",
    "-o", audio_path,
    YOUTUBE_URL
], check=True)

print("Downloaded:", audio_path)
```

```python
import os
import torch
from demucs.separate import main as demucs_main

def split_music_and_voice_mp3(
    audio_path,
    output_dir="output",
    bitrate=192
):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print("Using device:", device)

    os.makedirs(output_dir, exist_ok=True)

    demucs_main([
        "--two-stems", "vocals",
        "--mp3",
        "--mp3-bitrate", str(bitrate),
        "-d", device,
        "-o", output_dir,
        audio_path
    ])

    base = os.path.splitext(os.path.basename(audio_path))[0]
    stem_dir = os.path.join(output_dir, "htdemucs", base)

    vocals_mp3 = os.path.join(stem_dir, "vocals.mp3")
    music_mp3 = os.path.join(stem_dir, "no_vocals.mp3")

    return vocals_mp3, music_mp3
```

```python
vocals, music = split_music_and_voice_mp3(audio_path)
print("Generated files:")
print(vocals)
print(music)
```
